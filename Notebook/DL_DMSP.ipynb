### Baseine Model

```python
import tensorflow as tf
from tensorflow import keras
from keras import models ,layers
```

```python
b_model=[]
b_model = keras.Sequential()
b_model.add(layers.Dense(40, activation='tanh', kernel_initializer='normal',input_shape=(X_train.shape[1],)))
b_model.add(layers.Dense(30, activation='tanh', kernel_initializer='normal'))
#b_model.add(layers.Dropout(0.1))
b_model.add(layers.Dense(1, kernel_initializer='normal',activation='linear'))

b_model.summary()
```

```python
b_model.compile(optimizer='adam', loss='mean_absolute_error',metrics=['mean_absolute_error'])
```

```python
from keras.callbacks import ModelCheckpoint

ckpt_name = 'Weights-{epoch:03d}--{val_loss:.5f}.hdf5' 
ckpt = ModelCheckpoint(ckpt_name, monitor='val_loss', verbose = 1, save_best_only = True, mode ='auto')
callbacks_list = [ckpt]
```

```python
history = b_model.fit(X_train, y_train, epochs=500, batch_size=32, validation_split = 0.2, callbacks=callbacks_list, verbose=0)
```

```python
print(b_model.evaluate(X_train, y_train, verbose=0))
```

```python
b_model.load_weights("Weights-490--265803153408.00000.hdf5")
print(b_model.evaluate(X_train, y_train, verbose=0))
```

### Hyper-Tuning Parameters

```python
def model_builder(hp):
    model_fine_tuner=[]
    model_fine_tuner = keras.Sequential()
    hp_units = hp.Int('units', min_value=16, max_value=30, step=2)
    model_fine_tuner.add(layers.Dense(40, activation='tanh', kernel_initializer='normal',input_shape=(X_train.shape[1],)))
    model_fine_tuner.add(layers.Dense(30, activation='tanh', kernel_initializer='normal'))
    model_fine_tuner.add(keras.layers.Dense(1,activation='linear'))
    hp_lr = hp.Choice('learning-rate', values=[0.05,0.1,0.2,0.4])
    model_fine_tuner.compile(optimizer=keras.optimizers.Adam(learning_rate=hp_lr), loss='mean_absolute_error',metrics=['mean_absolute_error'])
    return model_fine_tuner
```

```python
import keras_tuner as kt
from keras_tuner import HyperModel

tune= kt.Hyperband(model_builder, max_epochs=10, objective='mean_absolute_error', seed=42, executions_per_trial=2, directory='dirf')
```

```python
tune.search_space_summary(extended=True) 
```

```python
stop_early = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)
tune.search(X_train, y_train, epochs=10, validation_split=0.3, callbacks=[stop_early] )
```

### Final Model

```python
from tensorflow import keras
import keras_tuner as kt

from keras.callbacks import ModelCheckpoint
from keras_tuner import HyperModel
```

```python
final_model = keras.Sequential()
final_model.add(layers.Dense(40, activation='tanh', kernel_initializer='normal',input_shape=(X_train.shape[1],)))
final_model.add(layers.Dense(28, activation='tanh', kernel_initializer='normal'))
final_model.add(keras.layers.Dense(1,activation='linear'))
final_model.compile(optimizer=keras.optimizers.Adam(learning_rate=0.75), loss='mean_absolute_error',metrics=['mean_absolute_error'])
```

```python
ckpt_name = 'NewWeights-{epoch:03d}--{val_loss:.5f}.hdf5' 
ckpt = ModelCheckpoint(ckpt_name, monitor='val_loss', verbose = 1, save_best_only = True, mode ='auto')
callbacks_list = [ckpt]
```

```python
history = final_model.fit(X_train, y_train, epochs=12, validation_split = 0.2, callbacks=callbacks_list, verbose=0)
```

```python
print(final_model.evaluate(X_train, y_train, verbose=0))
```

```python
final_model.load_weights("NewWeights-012--265793634304.00000.hdf5")
print(final_model.evaluate(X_train, y_train, verbose=0))
```
